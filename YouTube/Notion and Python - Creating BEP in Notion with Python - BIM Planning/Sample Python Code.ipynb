```python
import requests
import json

def create_notion_page():
    notion_api_key =  '' # Replace with your actual Notion API key
    parent_page_id = ''  # Replace with your actual Parent Page ID

    # User input section
    project_name = input("Enter the project name: ")
    revit_version = input("Enter the Revit version: ")
    links = input("Enter links related to the project (comma-separated): ")
    team_members = input("Enter team members (comma-separated): ")

    # Headers for the API request
    headers = {
        "Authorization": f"Bearer {notion_api_key}",
        "Content-Type": "application/json",
        "Notion-Version": "2022-06-28"
    }

    # Information content with line breaks
    project_info_content = f"Project Information:\n- Project Name: {project_name}\n- Revit Version: {revit_version}\n- Links: {links}\n- Team Members: {team_members}"

    # Constructing the data for the request
    data = {
        "parent": {"page_id": parent_page_id},
        "properties": {
            "title": {
                "title": [
                    {
                        "text": {
                            "content": project_name
                        }
                    }
                ]
            }
        },
        "children": [
            {
                "object": "block",
                "type": "paragraph",
                "paragraph": {
                    "rich_text": [
                        {
                            "type": "text",
                            "text": {
                                "content": project_info_content
                            }
                        }
                    ]
                }
            }
        ]
    }

    # API endpoint URL
    url = "https://api.notion.com/v1/pages"

    # Making the HTTP POST request to create a new page
    response = requests.post(url, headers=headers, data=json.dumps(data))

    # Verifying the response
    if response.status_code == 200:
        print("Page created successfully:", response.json())
    else:
        print(f"Failed to create page! Status code: {response.status_code}")
        print(response.text)

# Running the function
if __name__ == "__main__":
    create_notion_page()
```
